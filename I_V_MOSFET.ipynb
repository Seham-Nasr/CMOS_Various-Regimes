```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create vds vector
Vds = np.arange(0,5,0.01)
# create Vgs vector
Vgs = np.arange(0,5,1)
#Vdsat = Vgs -Vth
Vth = 0.7
Ids=np.zeros((500), dtype=float)
Beta= (120 * 2 )
for j in range(5):
    for i in range(500):
        if Vgs[j] < Vth:
            Ids [i] = 0
        if Vds[j] < (Vgs[j]-Vth):
            Ids [i] = Beta * (Vgs[j] - Vth - Vds[i]/2)* Vds[i]
        if Vds[i] > (Vgs[j]-Vth):
            Ids[i] = (Beta/2) * (Vgs[j] - Vth) ** 2
    plt.plot (Vds,Ids,label = ('Vgs='+str(Vgs[j])))
    
plt.title("MOSFET Ids-Vds characteristics")    
plt.xlabel(" Voltage Vds")
plt.ylabel('Current Ids')
#plt.grid(alpha=.4,linestyle='--')
plt.legend()
```
